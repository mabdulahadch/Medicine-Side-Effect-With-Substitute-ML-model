```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for better visualization
sns.set_style("whitegrid")

# Heatmap of missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cmap="viridis", cbar=True, yticklabels=False)
plt.title("Missing Values Heatmap", fontsize=14)
plt.show()

# Count of missing values per column
missing_counts = df.isnull().sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_counts.index, y=missing_counts.values, palette="magma")
plt.xticks(rotation=90)
plt.xlabel("Columns", fontsize=12)
plt.ylabel("Missing Values Count", fontsize=12)
plt.title("Missing Values Count per Column", fontsize=14)
plt.show()

# Distribution of numerical columns (if applicable)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
if len(numerical_cols) > 0:
    df[numerical_cols].hist(figsize=(12, 8), bins=50, edgecolor='black', color="skyblue")
    plt.suptitle("Numerical Column Distributions", fontsize=14)
    plt.show()
```

```python
def get_substitutes(medicine_name):
    result = df[df["name"].str.lower() == medicine_name.lower()]
    if result.empty:
        return "No substitutes found"
    return result.iloc[0][["substitute0", "substitute1", "substitute2", "substitute3","substitute4"]].dropna().tolist()

print(get_substitutes("augmentin 625 duo tablet"))
```
